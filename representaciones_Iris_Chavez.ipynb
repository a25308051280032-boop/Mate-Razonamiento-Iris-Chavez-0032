#### Problema 1
Paso 1: Explicación paso a paso

Problema 1: La suma de un número más 9 es igual a 21.

Planteamiento: Identificamos la incógnita. Llamemos $x$ al "número desconocido".

Traducción al lenguaje algebraico: "La suma de un número ($x$) más 9" se escribe como $x + 9$. "Es igual a 21" completa la ecuación:$$x + 9 = 21$$Despeje de la variable: Para dejar sola a la $x$, el 9 que está sumando pasa al otro lado de la igualdad restando:$$x = 21 - 9$$Solución: Realizamos la operación:$$x = 12$$El número es 12.

```python
#Problema 1: Encontrar el numero desconocido
# resultado_total = 21
# numero_conocido = 9
resultado_total =int(input("Ingrese el resultado total: "))
numero_conocido =int(input("Ingrese el numero conocido: "))

# Despejamos numero_incognito = resultado_total - numero_conocido
x = resultado_total - numero_conocido

print(f"Problema 1: El número buscado es {x}")
```

#### Problema 2
Paso 2: Explicación paso a paso

Problema 2:
Pedro tiene el doble de edad que su hermana Ana, y suman 36 años.

Planteamiento: Definimos las variables en función de una sola.Edad de Ana: $x$Edad de Pedro (el doble que Ana): $2x$

Traducción al lenguaje algebraico: "Entre los dos suman 36 años":$$x + 2x = 36$$Simplificación: Sumamos los términos semejantes ($1x + 2x$):$$3x = 36$$Despeje de la variable: El 3 que está multiplicando a la $x$ pasa al otro lado dividiendo:$$x = \frac{36}{3}$$Solución:$x = 12$ (Edad de Ana)$2x = 2 \cdot 12 = 24$ (Edad de Pedro)Verificación: $12 + 24 = 36$.

```python
# Solución Problema 2
# Ecuación: x + 2x = 36 -> 3x = 36 -> x = 36 / 3
# suma de edades = 36

suma_edades = 36
# La suma de las proporciones es 1 (Ana) + 2 (Pedro) = 3
proporcion_total = 3

# Calculamos la edad de Ana (x)
edad_ana = suma_edades / proporcion_total
# Calculamos la edad de Pedro (2x)
edad_pedro = edad_ana * 2

print(f"Problema 2: Ana tiene {int(edad_ana)} años y Pedro tiene {int(edad_pedro)} años.")
```

#### Problema 3:
Paso 3: Explicación paso a paso

Problema 3: Un número al que le reste 7 da como resultado 18. ¿Cuál es ese número?

Planteamiento: Identificamos la incógnita. Llamemos $x$ al "número desconocido".

Traducción al lenguaje algebraico: "Un número ($x$) al que le reste 7" se escribe como $x - 7$. "Da como resultado 18" completa la ecuación:$$x - 7 = 18$$Despeje de la variable: Para dejar sola a la $x$, el 7 que está restando pasa al otro lado de la igualdad sumando:$$x = 18 + 7$$Solución: Realizamos la operación:$$x = 25$$El número es 25.

```python
# Solución Problema 3
# x - 7 = 18
resultado_total = int(input("Ingrese el resultado total (18): "))
numero_conocido = int(input("Ingrese el numero conocido (7): "))

# Despejamos x = resultado_total + numero_conocido
x3 = resultado_total + numero_conocido

print(f"Problema 3: El número buscado es {x3}\n")
```

#### Problema 4:
Paso 4: Explicación paso a paso

Problema 4: El triple de un número más 4, es igual a 3. ¿Cuál será este número?

Planteamiento: Identificamos la incógnita. Llamemos $x$ al "número desconocido".

Traducción al lenguaje algebraico: "El triple de un número ($3x$) más 4" se escribe como $3x + 4$. "Es igual a 3" completa la ecuación:$$3x + 4 = 3$$Despeje de la variable: Primero, el 4 que suma pasa restando: $3x = 3 - 4$, lo que da $3x = -1$. Luego, el 3 que multiplica pasa dividiendo:$$x = \frac{-1}{3}$$Solución: Realizamos la operación:$$x \approx -0.33$$El número es -1/3 (o aproximadamente -0.33).

```python
# Solucion Problema 4:
# 3x + 4 = 3
resultado_total = int(input("Ingrese el resultado total (3): "))
numero_conocido = int(input("Ingrese el numero conocido (4): "))

# Despejamos x = (resultado_total - numero_conocido) / 3
x4 = (resultado_total - numero_conocido) / 3

print(f"Problema 4: El número buscado es {x4:.2f}\n")
```

#### Problema 5:
Paso 5: Explicación paso a paso

Problema 5: Un terreno de $5x + x + 5x + x$ tiene un perímetro total de 50 m. ¿Cuánto mide cada uno?

Planteamiento: Identificamos las variables. El ancho es $x$ y el largo es $5x$.

Traducción al lenguaje algebraico: La suma de todos los lados (perímetro) es igual a 50:$$5x + x + 5x + x = 50$$Despeje de la variable: Simplificamos sumando las $x$ para obtener $12x = 50$. Para dejar sola a la $x$, el 12 que multiplica pasa dividiendo:$$x = \frac{50}{12}$$Solución: Realizamos las operaciones:$x = 4.16$ (Ancho) y $5x = 20.83$ (Largo).El ancho mide 4.16 m y el largo 20.83 m.

```python
# Solucion Problema 5:
# 12x = 50
resultado_total = int(input("Ingrese el perímetro total (50): "))
numero_conocido = int(input("Ingrese la suma de las x (12): "))

# Despejamos x (ancho)
x5 = resultado_total / numero_conocido
# Calculamos el largo (5x)
largo = 5 * x5

print(f"Problema 5: El ancho mide {x5:.2f} m y el largo mide {largo:.2f} m")
```
