#### Problema 3
Paso 1: Explicación paso a pasoProblema 1: La suma de un número más 9 es igual a 21.Planteamiento: Identificamos la incógnita. Llamemos $x$ al "número desconocido".Traducción al lenguaje algebraico: "La suma de un número ($x$) más 9" se escribe como $x + 9$. "Es igual a 21" completa la ecuación:$$x + 9 = 21$$Despeje de la variable: Para dejar sola a la $x$, el 9 que está sumando pasa al otro lado de la igualdad restando:$$x = 21 - 9$$Solución: Realizamos la operación:$$x = 12$$El número es 12.

```python
#Problema 3: Encontrar el numero desconocido
# resultado_total = 21
# numero_conocido = 9
resultado_total =int(input("Ingrese el resultado total: "))
numero_conocido =int(input("Ingrese el numero conocido: "))

# Despejamos numero_incognito = resultado_total - numero_conocido
x = resultado_total - numero_conocido

print(f"Problema 3: El número buscado es {x}")
```

#### Problema 4
Problema 2: Pedro tiene el doble de edad que su hermana Ana, y suman 36 años.Planteamiento: Definimos las variables en función de una sola.Edad de Ana: $x$Edad de Pedro (el doble que Ana): $2x$Traducción al lenguaje algebraico: "Entre los dos suman 36 años":$$x + 2x = 36$$Simplificación: Sumamos los términos semejantes ($1x + 2x$):$$3x = 36$$Despeje de la variable: El 3 que está multiplicando a la $x$ pasa al otro lado dividiendo:$$x = \frac{36}{3}$$Solución:$x = 12$ (Edad de Ana)$2x = 2 \cdot 12 = 24$ (Edad de Pedro)Verificación: $12 + 24 = 36$.

```python
# Solución Problema 4
# Ecuación: x + 2x = 36 -> 3x = 36 -> x = 36 / 3
# suma de edades = 36

suma_edades = 36
# La suma de las proporciones es 1 (Ana) + 2 (Pedro) = 3
proporcion_total = 3

# Calculamos la edad de Ana (x)
edad_ana = suma_edades / proporcion_total
# Calculamos la edad de Pedro (2x)
edad_pedro = edad_ana * 2

print(f"Problema 4: Ana tiene {int(edad_ana)} años y Pedro tiene {int(edad_pedro)} años.")
```
